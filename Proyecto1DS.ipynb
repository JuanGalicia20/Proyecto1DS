# Se comienzan las transformaciones

Se convierte la columna de Area a booleana

```python
#Area
df[df["AREA"]=="SIN ESPECIFICAR"]["AREA"] = np.nan
df[df["AREA"]=="RURAL"]["AREA"] = True
df[df["AREA"]=="URBANA"]["AREA"] = False
df["AREA"]=df["AREA"].astype(bool)
print(df.dtypes)
```

```python
df
```

Se convierte la columna Status en booleana

```python
df.loc[df["STATUS"] == "ABIERTA", "STATUS"] = True
df.loc[df["STATUS"].isin(["CERRADA TEMPORALMENTE", "TEMPORAL TITULOS", "TEMPORAL NOMBRAMIENTO"]), "STATUS"] = False
df["STATUS"]=df["STATUS"].astype(bool)
```

Se convierte la columna Modalidad en booleana

```python
df.loc[df["MODALIDAD"] == "BILINGUE", "MODALIDAD"] = True
df.loc[df["MODALIDAD"]== "MONOLINGUE", "MODALIDAD"] = False
df["MODALIDAD"]=df["MODALIDAD"].astype(bool)
```

Se sintetizan las categorias en 3 para la columna Plan

```python
df.value_counts("PLAN")
planes = {'SEMIPRESENCIAL': 'SEMIPRESENCIAL/VIRTUAL',
          'SEMIPRESENCIAL (FIN DE SEMANA)': 'SEMIPRESENCIAL/VIRTUAL',
          'SEMIPRESENCIAL (UN DÃA A LA SEMANA)': 'SEMIPRESENCIAL/VIRTUAL',
          'A DISTANCIA': 'SEMIPRESENCIAL/VIRTUAL',
          'SEMIPRESENCIAL (DOS DÃAS A LA SEMANA)': 'SEMIPRESENCIAL/VIRTUAL',
          'VIRTUAL A DISTANCIA': 'SEMIPRESENCIAL/VIRTUAL',
          'INTERCALADO': 'SEMIPRESENCIAL/VIRTUAL',
          'IRREGULAR': 'SEMIPRESENCIAL/VIRTUAL',
          'FIN DE SEMANA': 'FIN DE SEMANA',
          'SABATINO': 'FIN DE SEMANA',
          'DOMINICAL': 'FIN DE SEMANA',
          'MIXTO': 'DIARIO(REGULAR)',
          'DIARIO(REGULAR)': 'DIARIO(REGULAR)'
          }
df['PLAN'] = df['PLAN'].map(planes)
```

Se investigo columna Director y se eliminan los caracteres no deseados y se convierten en nulos

```python
df["DIRECTOR"]
```

```python
df['DIRECTOR'] = df['DIRECTOR'].apply(lambda x: np.nan if isinstance(x, str) and '-' in x else x)
df["DIRECTOR"]
```

Se botÃ³ la columna Departamental ya que nos proveia la misma informacion que el Departamento

```python
df.drop("DEPARTAMENTAL", inplace=True, axis=1)
```

```python
df.dtypes
```

Se convirtio la columna de codigo a enteros, quitandole los caracteres -
esto para poder ordenarlo e identificar al mas reciente

```python
df["CODIGO"] = df["CODIGO"].str.replace("-","")
df["CODIGO"] = df["CODIGO"].astype(int)
```

```python
df
```

Se botan los duplicados que coincidan en estas 4 columnas, y se mantiene el mÃ¡s reciente

```python
df2 = df.sort_values(by="CODIGO", ascending=False)
df = df2.drop_duplicates(['DEPARTAMENTO', 'MUNICIPIO', 'ESTABLECIMIENTO', 'DIRECCION'])
df
```
